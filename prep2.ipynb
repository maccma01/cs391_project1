## Transforming categorical data into numerical data

```python
# Size column
```

```python
val_table = []
for value in df['size']:
    vals = value.split('-')
    for v in range(len(vals)):
        vals[v] = int(vals[v].strip("+").replace(",", ""))
    val_table.append(vals)
```

```python
count = 0
for v in val_table:
    if len(v) == 1:
        val_table[count] = v[0]
    else:
        val_table[count] = int(np.median([v[0], v[1]]))
    count+=1

df['size'] = val_table
```
